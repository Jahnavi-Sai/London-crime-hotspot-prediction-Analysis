```python
import pandas as pd

# Load dataset
df = pd.read_csv(r"C:\Users\91630\Downloads\london_crime.csv")

# Check data
print(df.head())
print(df.columns)
print(df.shape)
```

```python
# Convert wide month columns into rows
df_clean = df.melt(
    id_vars=["MajorText", "MinorText", "BoroughName"],
    var_name="Month",
    value_name="Crime_Count"
)
```

```python
# Convert Month from 202405 format to proper date
df_clean["Month"] = pd.to_datetime(df_clean["Month"], format="%Y%m")
```

```python
# Create extra columns for Power BI
df_clean["Year"] = df_clean["Month"].dt.year
df_clean["Month_Name"] = df_clean["Month"].dt.month_name()
df_clean["Month_Number"] = df_clean["Month"].dt.month
```

```python
# Rename columns
df_clean = df_clean.rename(columns={
    "MajorText": "Major_Crime_Type",
    "MinorText": "Minor_Crime_Type",
    "BoroughName": "Borough"
})
```

```python
# Remove missing values
df_clean = df_clean.dropna()
```

```python
# Save cleaned file
df_clean.to_csv("london_crime_cleaned_powerbi.csv", index=False)

print(df_clean.head())
print(df_clean.shape)
```
